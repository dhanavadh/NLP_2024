```python
import re
with open('corpus.txt', mode='w') as f:
  for line in open('TNC3_unicode_tokenized_all.txt'):
    line = re.sub('[\s\|]+', ' ',line).strip()
    if line != '':
      f.write(line)
      f.write('\n')
```
